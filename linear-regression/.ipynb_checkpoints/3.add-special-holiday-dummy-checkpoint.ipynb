```python
input_file ="sales_data.csv"
df = pd.read_csv(input_file)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Sales'] = pd.to_numeric(df['Sales'].str.replace(',', ''), errors='coerce').astype('int64')
df['Digital_Ads'] = pd.to_numeric(df['Digital_Ads'].str.replace(',', ''), errors='coerce').astype('int64')

# 変換後の情報を確認
df.info()
```

# Simple Regression Model

```python
# 目的変数 (Y): 売上数
Y = df['Sales']

# 説明変数 (X): TVCM, 紙媒体, OOH, WEB広告
X = df[['TVCM_GPR', 'Print_Media', 'Offline_Ads', 'Digital_Ads']]

# 定数項（切片）を追加
X = sm.add_constant(X)

# 重回帰分析のモデルを構築し学習する
model = sm.OLS(Y, X).fit()

# 分析結果の概要を表示する（補正 R² なども確認できる）
print(model.summary())

# オプション：残差プロットを作成して、モデルの適合度を視覚的にチェックする
plt.figure(figsize=(8, 4))
plt.scatter(model.fittedvalues, model.resid)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('予測値')
plt.ylabel('残差')
plt.title('残差プロット')
plt.show()
```

The model explains about 54% of the variance in Sales, which is a moderate level of explanatory power. Here are some insights based on the results:

1.Overall Model Fit
- R-squared: 0.539 (Adjusted R²: 0.521) indicates that roughly 54% of the variability in Sales is explained by the four predictors.
- F-statistic: 29.24 with a p-value of 4.24e-16 confirms that the overall model is highly significant.
    
2.Coefficient Insights:
- Intercept (4.55e+06): This represents the baseline Sales when all predictors are zero, around 4.55 million.
- TVCM_GPR (1561.01): For each one-unit increase in TVCM_GPR, Sales increase by about 1,561 units. However, this predictor is only marginally significant (p = 0.058), meaning its effect might not be as reliably estimated.
- Print_Media (0.0182): This variable is statistically significant (p = 0.005) and suggests that an increase in Print Media leads to a small, but significant, increase in Sales.
 - Offline_Ads (-0.0307): This predictor has a significant negative coefficient (p = 0.019). In this model, higher Offline Ads spending is associated with a decrease in Sales, which may be counterintuitive and worth investigating further.
 - Digital_Ads (32.06): Digital Ads have a strong, highly significant positive effect on Sales (p < 0.001), with an increase of about 32 units in Sales for each unit increase in Digital Ads.
    
3.Diagnostic Considerations:
 - Multicollinearity: The condition number is extremely high (3.98e+07), suggesting that multicollinearity may be an issue. This means some of the predictors might be highly correlated with each other, which can inflate the variance of the coefficient estimates.
 - Residuals: Tests like Omnibus and Jarque-Bera indicate that the residuals are not perfectly normally distributed, and the Durbin-Watson statistic (1.620) suggests some positive autocorrelation in the residuals. While these do not invalidate the model, they hint at potential areas for further investigation or model improvement.

Overall Insight:
 - While the model is statistically significant and explains over half of the variance in Sales, there are concerns about multicollinearity and non-normal residuals. 
 - Digital Ads and Print Media show robust positive effects, TVCM_GPR is marginally significant, and Offline Ads surprisingly show a negative effect. 
 - These findings suggest that while the predictors are useful, further model refinement—possibly addressing multicollinearity and reexamining the role of Offline Ads—is warranted for more reliable inference and prediction.

# Regression Model + Monthly Dummy

```python
import pandas as pd
import numpy as np

# Assume df already exists and has a "Date" column in datetime format
# If not, convert the Date column to datetime first:
df['Date'] = pd.to_datetime(df['Date'])

# Define a function that, given a week's start date, returns a Series with the count of days in each month.
def count_days_in_week(start_date):
    # Initialize a dictionary with keys "1月" to "12月" set to 0
    counts = {f'{i}月': 0 for i in range(1, 13)}
    # Assume the week spans 7 days
    for i in range(7):
        current_day = start_date + pd.Timedelta(days=i)
        month_str = f'{current_day.month}月'
        counts[month_str] += 1
    return pd.Series(counts)

# Define the list of dummy column names
dummy_columns = [f'{i}月' for i in range(1, 13)]

# Check if the dummy columns already exist in df; if not, create and concatenate them.
if not all(col in df.columns for col in dummy_columns):
    month_dummies = df['Date'].apply(count_days_in_week)
    df = pd.concat([df, month_dummies], axis=1)
else:
    print("Monthly dummy columns already exist. Skipping concatenation.")

print(df.tail(20))
```

```python
# # Save the processed DataFrame to a new CSV file, without including the index
# df.to_csv("processed_data.csv", index=False)
```

```python
# 目的変数 (Y): 売上数
Y = df['Sales']

# 説明変数 (X): TVCM, 紙媒体, OOH, WEB広告と月次ダミー変数
# 仮に、月次ダミー変数の列名が「2月」〜「12月」であるとします
dummy_cols = ['2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']
X = df[['TVCM_GPR', 'Print_Media', 'Offline_Ads', 'Digital_Ads'] + dummy_cols]

# 定数項（切片）の追加
X = sm.add_constant(X)

# 重回帰分析モデルの構築と学習
model = sm.OLS(Y, X).fit()

# 分析結果の概要を表示（補正R²なども確認可能）
print(model.summary())

# オプション：残差プロットでモデルの適合度を視覚的に確認
plt.figure(figsize=(8, 4))
plt.scatter(model.fittedvalues, model.resid)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('予測値')
plt.ylabel('残差')
plt.title('残差プロット')
plt.show()
```

```python

# If not already, store predicted and residual values in your DataFrame
df['Predicted'] = model.fittedvalues
df['Residual'] = model.resid

# Sort the DataFrame by date if needed
df.sort_values('Date', inplace=True)

# Create subplots: 
# Top subplot: Actual vs. Predicted 
# Bottom subplot: Residuals
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8), sharex=True)

# 1) Actual vs. Predicted
axes[0].plot(df['Date'], df['Sales'], label='Actual', marker='o')
axes[0].plot(df['Date'], df['Predicted'], label='Predicted', marker='x')
axes[0].set_ylabel('Sales')
axes[0].set_title('Actual vs. Predicted Sales')
axes[0].legend()

# 2) Residuals
axes[1].plot(df['Date'], df['Residual'], label='Residual', color='red', marker='o')
axes[1].axhline(0, color='black', linestyle='--')
axes[1].set_ylabel('Residual')
axes[1].set_xlabel('Date')
axes[1].set_title('Residual Plot')

plt.tight_layout()
plt.show()
```

The model explains roughly 54% of the variation in Sales (R² = 0.641, Adjusted R² = 0.580). Here are some key insights:
1. Baseline Sales:
The constant (≈ 4.47 million) represents the expected Sales when all predictors are zero. This is a significant baseline level.
2. Key Predictors:
- Digital_Ads: Has a strong positive effect (coef ≈ 34.03, p < 0.001). This indicates that, on average, each unit increase in Digital_Ads is associated with an increase of about 34 units in Sales.
- Offline_Ads: Shows a significant negative effect (coef ≈ -0.0373, p = 0.016). This negative coefficient is somewhat counterintuitive and might require further investigation or indicate an underlying issue (e.g., reverse causality or omitted variable bias).
- TVCM_GPR and Print_Media: Their coefficients (≈ 1534 and 0.0124, respectively) are marginally significant (p ≈ 0.069 and 0.064). This suggests a positive effect, but the evidence is not strong at the conventional 5% level.
3. Seasonality (Monthly Dummies):
- The monthly dummy variables (from “2月” to “12月”) largely show non-significant effects. 
- This might imply that once these advertising variables are accounted for, there isn’t a strong or consistent seasonal pattern in the data. 
- Alternatively, it could indicate redundancy or multicollinearity issues among the dummies.
4. Diagnostic Considerations:
- Multicollinearity: The condition number is extremely high (≈ 7.05e+07), suggesting that some predictors (possibly the monthly dummies) are highly correlated. This can inflate the standard errors of the coefficients, making it harder to detect significant effects.
- Residual Diagnostics: The Omnibus and Jarque-Bera tests indicate that the residuals deviate somewhat from normality (with skewness of 0.355 and kurtosis of 5.976), though the Durbin-Watson statistic (1.909) suggests that autocorrelation is not a major concern.

Overall Insight:
- While the model is statistically significant overall (F-statistic p < 0.001) and explains a moderate portion of the variance in Sales, some predictors (especially the monthly dummies) do not contribute significantly and multicollinearity is a concern. 
- Digital_Ads emerge as a robust positive driver, whereas Offline_Ads unexpectedly show a negative effect. It may be beneficial to revisit variable selection (perhaps reducing the number of seasonal dummies or combining highly correlated ones) and explore potential reasons behind the negative effect of Offline_Ads for improved model stability and interpretability.

# Regression Model + Monthly Dummy + Public Holiday Dummy

```python
from datetime import datetime, timedelta

# 1) Define the holiday date ranges for each year (adjust as needed).
#    Here, each tuple is (start_date, end_date).
#    We'll assume 2015-2017 for demonstration. Extend for more years as needed.
holiday_ranges = {
    '正月': [
        (datetime(2015,12,28), datetime(2016,1,3)),
        (datetime(2016,12,26), datetime(2017,1,3)),
        (datetime(2017,12,25), datetime(2018,1,3)),  # Extend if needed
    ],
    'GW': [
        (datetime(2015,4,29), datetime(2015,5,5)),
        (datetime(2016,4,29), datetime(2016,5,5)),
        (datetime(2017,4,29), datetime(2017,5,5)),
    ],
    'お盆': [
        (datetime(2015,8,13), datetime(2015,8,16)),
        (datetime(2016,8,13), datetime(2016,8,16)),
        (datetime(2017,8,13), datetime(2017,8,16)),
    ]
}

# 2) Create columns for the three dummies, initialized to 0.
df['正月'] = 0
df['GW'] = 0
df['お盆'] = 0

# 3) Define a function to check if a given 7-day window intersects a holiday range.
def intersects_holiday(week_start, holiday_start, holiday_end):
    """
    Returns True if the 7-day window [week_start, week_start+6] intersects
    with [holiday_start, holiday_end].
    """
    week_end = week_start + pd.Timedelta(days=6)
    return not (week_end < holiday_start or week_start > holiday_end)

# 4) Iterate over each row in df, check each holiday range, and set the dummy if any range intersects.
for idx, row in df.iterrows():
    week_start = row['Date']
    
    # Check each holiday type (正月, GW, お盆)
    for holiday_name, ranges in holiday_ranges.items():
        for (start_date, end_date) in ranges:
            if intersects_holiday(week_start, start_date, end_date):
                df.loc[idx, holiday_name] = 1
                break  # If it intersects one range in that holiday_name, no need to check further

# 5) Now you can add monthly dummies as before, or they might already exist in your df.
# Example: If you have monthly dummies from "1月" to "12月", they can remain as is.

# Display the first few rows to see the new holiday columns
print(df.head(15))

```
