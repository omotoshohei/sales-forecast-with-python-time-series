```python
import numpy as np
import pandas as pd
import datetime


from prophet import Prophet

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import TimeSeriesSplit

from pmdarima.model_selection import train_test_split

import optuna

import warnings
warnings.simplefilter('ignore')

import matplotlib.pyplot as plt
plt.style.use('ggplot') #グラフスタイル
plt.rcParams['figure.figsize'] = [12, 9] #グラフサイズ
plt.rcParams['font.size'] = 14 #フォントサイズ
```

```python
dataset ="../data/sales_data.csv"
df = pd.read_csv(dataset)
```

# Preprocess

```python
# Clean numeric columns
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y')
df['Sales'] = pd.to_numeric(df['Sales'].str.replace(',', ''), errors='coerce').fillna(0).astype('int64')

df.info()
```

```python
# Filter out Actual records for model building
df_actual = df[df['Record_Type'] == "Actual"].copy()
# Load Forecast data (rows with Record_Type "Forecast")
df_forecast = df[df['Record_Type'] == "Forecast"].copy()
```

```python
# Filter out Actual records for model building
df_actual = df_actual.reset_index()[["Date", "Sales"]]
# Load Forecast data (rows with Record_Type "Forecast")
df_forecast = df_forecast.reset_index()[["Date", "Sales"]]
```

```python
# =========================================================================
# 1. データ読み込み・前処理
# =======================================================================
# Rename columns for clarity
df_actual = df_actual.rename(columns={"Date": "ds", "Sales": "y"})
```

```python
df_actual
```

# Test Train Split

```python
# =========================================================================
# 2. 学習データとテストデータに分割
#    (ここでは直近10行をテストデータとして分割しています)
# =========================================================================

train, test = train_test_split(df_actual, test_size=12)

```

```python
train
```

# Visualize the Test / Train Data

```python
# 確認用のグラフ（学習データとテストデータ）
fig, ax = plt.subplots()

# 学習データの描写
ax.plot(
    train.index,
    train.y.values,
    linestyle='-',
    label='train'
)

# テストデータの描写
ax.plot(
    test.index,
    test.y.values,
    linestyle='--',
    label='test',
    color='gray'
)

# 学習データとテスデータの間の縦線
ax.axvline(
    94,
    color='blue')
ax.legend()
plt.show()
```

# Model Building

```python
# =========================================================================
# 3. Prophet モデルの作成・学習
#    └─ ハイパーパラメータを固定し、ds/y のみで予測
# =========================================================================

# シンプルなモデル (デフォルト設定)
prophet = Prophet(weekly_seasonality=True)

# 学習
prophet.fit(train)
```

```python
# =========================================================================
# 4. 予測の実施
# =========================================================================

# 予測は学習データ期間＋テストデータ期間(10週間分)
df_future = prophet.make_future_dataframe(
    periods=12,
    freq='W-MON'   # 週次（月曜）
)
df_pred = prophet.predict(df_future)
fig = prophet.plot(df_pred)
# a = add_changepoints_to_plot(fig.gca(), prophet, df_pred)
```

```python
# =========================================================================
# 5. 元のデータに予測値を結合して誤差計算
# =========================================================================
df_actual['predicted'] = df_pred['yhat']

# 予測値を学習期間とテスト期間に分ける
train_pred = df_actual.iloc[:-12]['predicted']  # 学習データ期間
test_pred = df_actual.iloc[-12:]['predicted']   # テストデータ期間

# テストデータの予測精度
print('RMSE: ', np.sqrt(mean_squared_error(test.y, test_pred)))
print('MAE:  ', mean_absolute_error(test.y, test_pred))
print('MAPE: ', mean_absolute_percentage_error(test.y, test_pred))
```

```python
# =========================================================================
# 6. 予測値と実測値を可視化
# =========================================================================
fig, ax = plt.subplots()

# 実測値
ax.plot(train.index, train.y, linestyle='-', label='actual(train)')
ax.plot(test.index,  test.y,  linestyle='--', label='actual(test)', color='gray')

# 予測値
ax.plot(train.index, train_pred, linestyle=':', color='c')
ax.plot(test.index,  test_pred,  linestyle=':', label='predicted', color='c')

ax.axvline(94, color='blue')  # 学習とテストを区切る縦線
ax.legend()
plt.show()
```
