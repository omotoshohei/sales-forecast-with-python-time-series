# Introduction

## Sales Forecasting Model Using Prophet

This repository contains a base sales forecasting model built in Python using the Prophet library. The model is designed to forecast weekly sales data by capturing trends and seasonal patterns inherent in the data. In this implementation, Prophet is used to automatically detect changepoints and seasonality, while also allowing the inclusion of external regressors (such as key advertising metrics) to enhance forecast accuracy.

◼️ Model Building
- Prophet Model:
A base Prophet model is constructed with built-in functionalities for trend and seasonality detection.
- External Regressors:
Key advertising variables are added as regressors to allow the model to account for their impact on sales. These regressors are integrated before model fitting, ensuring that their influence is captured in the forecast.

◾️ Model Evaluation
- Train/Test Split:
The “Actual” sales data is split into training and testing sets. The model is trained on the training set and then evaluated on the test set using metrics such as RMSE, MAE, and MAPE.

◾️ Forecasting
- Future Predictions:
After validating the model, it is applied to forecast future sales for a specified period. The forecasted values are output and visualized alongside the historical data.

◾️ Overall Insight
- Although the predictive accuracy of this base model may be moderate, it provides a solid framework for understanding time series forecasting using Prophet. 
- The model’s structure, including the integration of external regressors, offers a clear pathway for further refinement and enhancement in future iterations.

```python
import numpy as np
import pandas as pd
import datetime


from prophet import Prophet

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import TimeSeriesSplit

from pmdarima.model_selection import train_test_split

import optuna

import warnings
warnings.simplefilter('ignore')

import matplotlib.pyplot as plt
plt.style.use('ggplot') #グラフスタイル
plt.rcParams['figure.figsize'] = [12, 9] #グラフサイズ
plt.rcParams['font.size'] = 14 #フォントサイズ
```

```python
dataset ="../data/sales_data.csv"
df = pd.read_csv(dataset)
```

# Preprocess

```python
# Clean numeric columns
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y')
df['Sales'] = pd.to_numeric(df['Sales'].str.replace(',', ''), errors='coerce').fillna(0).astype('int64')
df['Digital_Ads'] = pd.to_numeric(df['Digital_Ads'].str.replace(',', ''), errors='coerce').fillna(0).astype('int64')
df['Offline_Ads'] = pd.to_numeric(df['Offline_Ads'].astype(str).str.replace(',', ''), errors='coerce').fillna(0).astype('int64')
df['TVCM_GPR'] = pd.to_numeric(df['TVCM_GPR'].astype(str).str.replace(',', ''), errors='coerce').fillna(0)
df['Print_Media'] = pd.to_numeric(df['Print_Media'].astype(str).str.replace(',', ''), errors='coerce').fillna(0)

df.info()
```

```python
# Filter out Actual records for model building
df_actual = df[df['Record_Type'] == "Actual"].copy()
# Load Forecast data (rows with Record_Type "Forecast")
df_forecast = df[df['Record_Type'] == "Forecast"].copy()
```

```python
# Filter out Actual records for model building
df_actual = df_actual.reset_index()[["Date", "Sales","Digital_Ads", "Offline_Ads","TVCM_GPR", "Print_Media"]]
# Load Forecast data (rows with Record_Type "Forecast")
df_forecast = df_forecast.reset_index()[["Date", "Sales","Digital_Ads", "Offline_Ads","TVCM_GPR", "Print_Media"]]
```

```python
# =========================================================================
# 1. データ読み込み・前処理
# =======================================================================
# Rename columns for clarity
df_actual = df_actual.rename(columns={"Date": "ds", "Sales": "y"})
df_forecast = df_forecast.rename(columns={"Date": "ds", "Sales": "y"})
```

```python
df_actual
```

# Test Train Split

```python
# =========================================================================
# 2) Train/Test Split (for evaluation) on df_actual
# =========================================================================
train, test = train_test_split(df_actual, test_size=12)

# Print shapes just for clarity
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
train
```

# Visualize the Test / Train Data

```python
# 確認用のグラフ（学習データとテストデータ）
fig, ax = plt.subplots()

# 学習データの描写
ax.plot(
    train.index,
    train.y.values,
    linestyle='-',
    label='train'
)

# テストデータの描写
ax.plot(
    test.index,
    test.y.values,
    linestyle='--',
    label='test',
    color='gray'
)

# 学習データとテスデータの間の縦線
ax.axvline(
    94,
    color='blue')
ax.legend()
plt.show()
```

# Model Building

```python
# =========================================================================
# 3) Prophet Model with External Regressors
# =========================================================================
model = Prophet(weekly_seasonality=True)

# Add regressor columns
model.add_regressor("Digital_Ads")
model.add_regressor("Offline_Ads")
model.add_regressor("TVCM_GPR")
model.add_regressor("Print_Media")

# Fit on train
model.fit(train)
```

```python
# =========================================================================
# 4) Predict on Test Set (Evaluation)
# =========================================================================
# For Prophet, we can supply the test rows directly, since they have ds + regressor columns
test_pred_df = model.predict(test)
test_pred = test_pred_df['yhat'].values

rmse = np.sqrt(mean_squared_error(test['y'], test_pred))
mae = mean_absolute_error(test['y'], test_pred)
mape = mean_absolute_percentage_error(test['y'], test_pred)
print(f"\n--- Test Evaluation ---")
print("RMSE:", rmse)
print("MAE: ", mae)
print("MAPE:", mape)
```

```python
# =========================================================================
# 5) Forecast for Future Data (df_forecast)
# =========================================================================
# Suppose df_forecast also has ds plus the same regressor columns for future dates
# Then we can call model.predict(...) directly:
forecast_pred_df = model.predict(df_forecast)

print("\n--- Forecast (Future) Results ---")
print(forecast_pred_df[['ds', 'yhat', 'yhat_lower', 'yhat_upper']])
```

```python
# =========================================================================
# 6) Plotting
# =========================================================================
fig, ax = plt.subplots()

# (a) Plot train actual
ax.plot(train['ds'], train['y'], label='Train Actual', color='gray')

# (b) Plot test actual
ax.plot(test['ds'], test['y'], label='Test Actual', linestyle='--', color='gray')

# (c) Plot test predictions
ax.plot(test['ds'], test_pred, label='Test Predicted', linestyle=':', color='blue')

# (d) Plot forecast (future) predictions
ax.plot(forecast_pred_df['ds'], forecast_pred_df['yhat'], label='Future Forecast', linestyle='--', color='red')

ax.set_xlabel("Date")
ax.set_ylabel("Sales")
ax.set_title("Prophet Model with Regressors: Train/Test/Forecast")
ax.legend()
plt.show()
```

# Result

These metrics show how well (or poorly) the model performed on the held-out test set:

- RMSE (773,890): On average, the model’s predictions deviate from the actual sales by about 774k (in whatever sales units you have). A higher RMSE means large individual errors have more impact on the final score.
- MAE (486,126): On average, the absolute difference between predicted and actual sales is around 486k. This is a more direct measure of the average size of errors.
- MAPE (6.57%): The model’s predictions are, on average, within about 6.6% of the true sales values—this is often a key indicator of relative forecasting accuracy.

In short, the MAPE of ~6.6% suggests the model is reasonably accurate relative to the scale of sales, while the absolute errors (RMSE, MAE) remind us that forecasts can still be off by hundreds of thousands in absolute terms. 

Whether this performance is “good” depends on the business context—if a 6–7% error margin is acceptable for planning and decision-making, the model may be sufficient. Otherwise, you might explore additional features, more advanced modeling, or extended historical data to reduce error further.
